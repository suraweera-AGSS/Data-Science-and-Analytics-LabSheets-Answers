```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
```

```python
zoo_data = pd.read_csv("zoo.csv")
```

### split dataset into 2 sets of data using iloc method

```python
x = zoo_data.iloc[:,1:17]
```

```python
x.shape
```

```python
x.head()
```

```python
y = zoo_data.iloc[:,17]
```

```python
y.shape
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.head()
```

```python
x_test.shape
```

```python
y_train.head()
```

```python
y_train.shape
```

```python
y_test.head()
```

```python
y_test.shape
```

```python
zoo_classifier = DecisionTreeClassifier(random_state=0) # Building the classifier called "zoo_classifier"
```

```python
zoo_classifier.fit(x_train,y_train) # Train the classifier using training data
```

```python
zoo_classifier.score(x_test,y_test) # Cheacking the accuracy of classifier using testing data
```

```python
# accuracy > 0.75 ; It's completely fine model which means DecisionTreeClassifier fixed the secnario well
# if not this is not a good model
```
