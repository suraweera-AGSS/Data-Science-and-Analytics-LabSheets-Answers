```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
```

```python
import statsmodels.api as sm 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("salary_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:,:-1].values
```

```python
x
```

```python
y = df.iloc[:,1].values
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
model = LinearRegression() #create a regressor
```

```python
model.fit(x,y)
```

```python
model.predict(np.array([[5]]))
```

```python
x1 = sm.add_constant(x)
```

```python
model = sm.OLS(y,x1).fit() 
model.summary()
```
